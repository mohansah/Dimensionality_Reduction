# 2D Visualization using PCA

```python
#Considering only 15000 of data-points

data=x_train.head(15000)
labels=y_train.head(15000)

print(data.shape)
print(labels.shape)
```

```python
data.head(20)
```

```python
# Data-Preprossing : Standardizing the data

from sklearn.preprocessing import StandardScaler

std_data=StandardScaler().fit_transform(data)
std_data
```

```python
std_data.shape
```

```python
# finding co-variance matrix i.e., X^T * X

X=std_data
print(X.shape)

#matrix multiplication using numpy
co_var_mat=np.matmul(X.T,X)

print(co_var_mat.shape)
```

```python
co_var_mat
```

```python
#Finding eigen values and eigen vectors
#for projecting onto a 2d space

from scipy.linalg import eigh

# the parameter 'eigvals' if defined (low value to high value)
# eigh function will return the eigen values in asending order
# this code generates only the top 2 (782 and 783) eigen values

values,vectors=eigh(co_var_mat,eigvals=(782,783))
print(vectors.shape)
```

```python
values
```

```python
vectors
```

```python
#Converting eigen vectors from 784X2 t0 2X784

vectors=vectors.T
print(vectors.shape)

#vectors[1] represent the eigen vector corresponding 1st principal component
#vectors[0] represent the eigen vector corresponding 2nd principal component
vectors
```

```python
#projecting the original data sample on the plane
#formed by two principal eigen vectors by vector vector multiplication

new_cor=np.matmul(vectors,X.T)     #(2,784) X (784,15000) = (2,15000)
new_cor.shape
```

```python
# appending label to the projected data
new_cor=np.vstack((new_cor,labels)).T
new_cor
```

```python
#creating a new data frame for plotting the labeled points

df_f=pd.DataFrame(data=new_cor,columns=('PC2','PC1','Label'))
df_f.head()
```

```python
df_f.shape
```

```python
#ploting the 2nd data points with seaborn

import seaborn as sn

sn.FacetGrid(df_f,hue='Label',size=15).map(plt.scatter,'PC1','PC2').add_legend()
plt.show()
```

# PCA using Scikit-Learn

```python
# we have standarized data in std_data

std_data
```

```python
from sklearn.decomposition import PCA

pca_data=PCA(n_components=2).fit_transform(std_data)
pca_data
```

```python
pca_data.shape
```

```python
pca_data=np.vstack((pca_data.T,labels)).T
pca_data
```

```python
pca_df=pd.DataFrame(data=pca_data,columns=('PC1','PC2','Label'))
pca_df.head()
```

```python
#ploting the 2nd data points with seaborn

import seaborn as sn

sn.FacetGrid(pca_df,hue='Label',size=15).map(plt.scatter,'PC1','PC2').add_legend()
plt.show()
```

```python
# want to acheive 95% of the variance

pca1=PCA(.95)
pca1_d=pca1.fit_transform(std_data)
pca1_d
```

```python
pca1_d.shape
```

```python
print("Number of components req for 95% of the variance : ",pca1.n_components_)
```

# Explained Variance

```python
pca_n=PCA(n_components=784)     #considering all 784 components
pca_data_n=pca_n.fit_transform(std_data)         
per_var_explained=pca_n.explained_variance_ratio_      # or pca_n.explained_variance_/np.sum(pca_n.explained_variance_)
per_var_explained
```

```python
cumulative_var_explained=np.cumsum(per_var_explained)
cumulative_var_explained
```

```python
plt.figure(figsize=(15,12))
plt.grid()
plt.plot(range(1,785),cumulative_var_explained,c='r')
plt.xlabel('Number of Components',fontsize=15)
plt.ylabel('Cumulative Explained Variance',fontsize=15)
plt.show()
```

If we take 200-dimensions, approx 90% variance is explained and if we take 600-dimensions, approx 100% variance is explained

# t-SNE using scikit-learn

```python
print(std_data.shape)
std_data
```

```python
print(labels.shape)
labels
```

```python
#Taking 1000 data points
data_1000=std_data[0:1000]
labels_1000=labels[0:1000]
print(data_1000.shape)
print(labels_1000.shape)
```

```python
from sklearn.manifold import TSNE

model=TSNE(n_components=2,random_state=0)   # random_state=0, make sure that same result is produced on running multiple times
#default perplexity=30
#default learnig rate=200
#default maximum iterations for the optimization=1000

tsne_data=model.fit_transform(data_1000)      #Fit data_1000 into an embedded space and return that transformed output
tsne_data
```

```python
#creating new dataframe
tsne_df=pd.DataFrame(data=tsne_data,columns=('DM1','DM2'))
tsne_df.head()
```

```python
tsne_df['Label']=labels_1000
tsne_df.head()
```

```python
#plotting

sn.FacetGrid(tsne_df,hue='Label',size=10).map(plt.scatter,'DM1','DM2').add_legend()
plt.title('With perplexity=30 and iterations=1000')
plt.show()
```
